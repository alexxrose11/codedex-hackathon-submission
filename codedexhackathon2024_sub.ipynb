```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('bmh')   

# Data Set 1 - Tokyo 2021 Olympics
df = pd.read_csv(r"C:\Users\alexa\Documents\tokyo_2021_medals_dataset.csv")
```

```python
team_noc = df['Team/NOC'].values
total_num_of_medals = df['Total'].values

sorted_indices = np.argsort(total_num_of_medals)[::-1]
sorted_team_noc = team_noc[sorted_indices]
sorted_total_num_of_medals = total_num_of_medals[sorted_indices]
top_10_percent_threshold = int(0.1 * len(sorted_team_noc))

# x and y for Data Set 1
x = team_noc[:top_10_percent_threshold]
y = total_num_of_medals[:top_10_percent_threshold]
```

```python
# Pie Chart for Tokyo 2021 Olympics
plt.title("Top 10% of Countries with Most Total Medals (Tokyo 2021 Olympics)", y=1.1, fontdict={'family': 'sans-serif'})
colors = ['#74c49b', '#e876df', '#cc6127', '#a676f5', '#99466a', '#2fdedb', '#2f32de', '#e3cb6d', '#842dcc']
explode = [0.1 if i==0 else 0 for i in range(len(x))] # Explode U.S. Slice
plt.pie(y, labels=x, radius=1.2, autopct='%1.1f%%', shadow=True, colors=colors, explode=explode)
# Text with my hypothesis
plt.text(2, 0, 'The U.S.A. is most likely to win the highest number of total medals \nbased off the results of the Tokyo 2021 Olympics',
         color='green', ha='center', fontsize=14, bbox=dict(facecolor='white', alpha=0.5), fontdict={'family': 'fantasy'})
plt.show()
```

```python
# Data Set 2 - London 2012 Olympics
df2 = pd.read_csv(r"C:\Users\alexa\Documents\rio_olympics_2012_medals.csv")
```
